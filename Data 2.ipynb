```python
# Specify the file path
file_path = 'mediterranean-data(2024-04-11).xlsx'

# Read the Excel file into a DataFrame
df = pd.read_excel(file_path)
```

```python
df
```

```python
arrivals_df = df[df["field_activity"] == "Irregular arrivals in Europe"]
```

```python
arrivals_df
```

```python
arrivals_df["field_count"].sum()
```

```python
# Step 1: Sum of field_count
total_field_count = arrivals_df["field_count"].sum()
print("Total field count:", total_field_count)

# Grouping by year and month and summing the field_count
grouped_data = arrivals_df.groupby(['Year', 'field_month'])['field_count'].sum().reset_index()

# Displaying the grouped data
grouped_data
```
